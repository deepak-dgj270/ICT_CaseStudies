# 1. Loading the dataset
We are reading the csv dataset into the variable 'loanData'. We are converting the loanData into a pandas DataFrame.

```python
loanData = pd.read_csv("train_loan.csv")
DF_loanData = pd.DataFrame(loanData)
```

# 2a. Missing Value Handling

# 2a.1 fillna()
This method will fill replace NaN values with the value we provide

```python
#filling NaN with 0
FilledDF = DF_loanData.fillna(0)

FilledDF
```

# 2a.2 Filling null values with previous row value
This function replaces NaN with the previous row value

```python
FilledDF = DF_loanData.fillna(method = 'pad')
FilledDF
```

# 2a.3 Filling null values with next row

This replaces NaN with the next row value

```python
FilledDF = DF_loanData.fillna(method = 'bfill')
FilledDF
```

# 2b. Outlier Detection

```python
#creating a box plot based on the loan data csv with 'loan amount' as the
plt.boxplot(FilledDF['LoanAmount'])
plt.title("BoxPlot of Loan Data based on Loan Amount")
plt.show()
```

```python
#Creating a function to create boxplots of multiple columns

def BPCreator(columnName):
  plt.boxplot(FilledDF[f"{columnName}"])
  plt.title("BoxPlot of Loan Data based on "+columnName)
  plt.show()

BPCreator('ApplicantIncome')
BPCreator('CoapplicantIncome')
BPCreator('Loan_Amount_Term')
```

# 2c. Encoding

```python
#Label Encoding
#convert ategorical column into a numerical colum

#imprt LabelEncoder library from sklearn to preprocess the data using label encoder
from sklearn.preprocessing import LabelEncoder

#Use Label Encoder to provide numberical values to the column 'Property Area'
le = LabelEncoder()
FilledDF['Property_Area'] = le.fit_transform(FilledDF['Property_Area'])
FilledDF
```

```python
#One Hot Encoding
#Enocodes just the Gender column
onehotDF = pd.get_dummies(FilledDF,columns = ['Gender'], dtype = int)
onehotDF
```

# 2d. Scaling

```python
#importing library
from sklearn.preprocessing import MinMaxScaler

#Using min-max scaler to assign the column values between range 1 and 0
min_max = MinMaxScaler()
FilledDF.loc[:,['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term']] = min_max.fit_transform(FilledDF[['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term']])
FilledDF
```
